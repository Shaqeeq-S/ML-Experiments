SCENARIO 1: E-commerce Sales Data
Analyze a sample e-commerce dataset to understand product categories, sales trends, and missing information.
Dataset (Kaggle – Public): 
https://www.kaggle.com/datasets/carrie1/ecommerce-data

IN-LAB TASKS:
•	Import necessary Python libraries.
•	Load the dataset into a Pandas DataFrame.
•	Inspect dataset using head(), tail(), info(), and describe().
•	Check for missing values.
•	Visualize sales per product using bar and line charts.


```python
# Load the dataset into a Pandas DataFrame.
filepath = "C:\\Users\\HP\\Downloads\\archive(1)\\data.csv"
df = pd.read_csv(filepath)
```

```python
# Inspect dataset using head(), tail(), info(), and describe().
print("First Five Rows:\n", df.head(5))
print("\nLast Five Rows:\n", df.tail(5))
print("\nDataFrame Info:")  
print(df.info())
print("\nStatistical Summary:\n", df.describe())
```

```python
# Check for missing values.
print("Missing values from dataset")
print(df.isnull().sum())
```

```python
# Visualize sales per product using bar and line charts.
df["Sales"] = df["Quantity"] * df["UnitPrice"]
sales_per_product = (
    df.groupby("Description")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
print(sales_per_product.index)
print(sales_per_product.values)
plt.figure(figsize=(20,10))
plt.bar(sales_per_product.index, sales_per_product.values, color='skyblue')
```

```python
# line chart
plt.figure(figsize=(20,10))
plt.plot(sales_per_product.index, sales_per_product.values, marker='o', color='orange')
plt.xticks(rotation=45)
plt.xlabel("Product Description")
plt.ylabel("Total Sales")
plt.title("Top 10 Products by Sales")
plt.grid()
```

Interpret trends for price prediction (Housing Dataset)

When analyzing a housing dataset, the following key trends help in predicting house prices:

Area vs Price (Positive Trend)
House price generally increases as area increases, indicating that area is a strong predictor of price.

Number of Rooms (Bedrooms/Bathrooms)
Houses with more bedrooms and bathrooms tend to have higher prices, showing a direct relationship with property value.

Location-Based Features
Properties located in preferred areas or with amenities (e.g., main road access, guest room, air conditioning) usually have higher prices.

Feature Correlation
Correlation heatmaps often show that area, bathrooms, and parking have strong positive correlation with price, making them important features for prediction models.

Diminishing Returns
After a certain point, increasing size or features results in smaller price increases, indicating non-linear behavior.

```python
plt.figure()
plt.hist(BankData["age"], bins=20)
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.title("Age Distribution of Bank Customers")
plt.show()


# 2. Balance Distribution (Box Plot)

plt.figure()
sns.boxplot(y=BankData["balance"])
plt.ylabel("Account Balance")
plt.title("Account Balance Distribution")
plt.show()


# 3. Spending / Interaction Pattern (Duration Box Plot)

plt.figure()
sns.boxplot(y=BankData["duration"])
plt.ylabel("Call Duration (seconds)")
plt.title("Customer Interaction Duration")
plt.show()




```

Age Distribution: The customer base is primarily middle-aged, with most customers clustered between roughly 30–55 years. Very young and very old customers form a smaller proportion, indicating the bank’s core users are working professionals.

Income Pattern: Income distribution is right-skewed. Most customers fall into low-to-medium income brackets, while a small number of high-income customers appear as outliers in box plots. This suggests income inequality within the customer base.

Spending Behavior: Spending patterns vary significantly across customers. Box plots show several high-spending outliers, indicating a small segment of customers contributes disproportionately to total spending.

Income vs Spending Insight: Customers with higher income generally show higher spending, but the relationship is not perfectly linear—some mid-income customers also exhibit high spending, suggesting lifestyle or credit usage effects.

Data Quality Observation: Missing values (especially in income) are visible and can affect interpretation. These should be handled before building predictive or segmentation models.
